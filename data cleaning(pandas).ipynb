```python
import pandas as pd 
import numpy as np
import os
```

```python
data = pd.read_csv("pandas_data1.csv")
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#filling null values in age column
avg_age = data['Age'].mean()
```

```python
avg_age
```

```python
data['Age'] = data.Age.fillna(avg_age)
```

```python
data['Age'] = data.Age.round(decimals=1)
```

```python
data.Age
```

```python
data.Rating
```

```python
avg_rating = data['Rating'].mean()
```

```python
avg_rating
```

```python
data['Rating'] = data.Rating.fillna(avg_rating)
```

```python
data['Rating'] = data.Rating.round(decimals=1)
```

```python
data.Rating
```

```python
data.Salary
```

```python
data['Salary'] = data['Salary'].str.replace('$',"").str.replace('k',"000")
```

```python
data.Salary
```

```python
data.Location
```

```python
data['Location'] = data['Location'].str.replace("India,In","India").replace("New York,Ny","New York").replace("Australia Aus","Australia").replace("India In","India")
```

```python
data.Location
```

```python
data['Established'].unique()
```

```python
data['Established'] = np.where(data['Established'] >= 0, data['Established'], pd.NA)
```

```python
data.Established
```

```python
data
```

```python
data.columns = data.columns.str.replace(" ","_")
```
